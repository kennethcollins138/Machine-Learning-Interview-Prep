# Intuition

This concept is pretty similar to linear regression! Think about back in school when we had linear equations such as y = ax^2 + bc + c. 
The powered terms introduced curved relationships to our "straight lines" from simple and multiple linear regression.

Polynomial regression is just an extension of the concepts I talked about in linear but to higher degrees! Instead of just having terms like `β1*x` (which gives a straight line) or `β1*x + β2*x^2` (which gives a parabola), you can have terms like `β1*x + β2*x^2 + β3*x^3 + ... + βn*x^n`. This allows you to model relationships that curve in complex ways. Much like linear regression, polynomial regression has an error term to handle the differences since our model won't be 100% accurate.

**Important note:**
- As we increase the degree of polynomial, it becomes super easy to over fit as our model becomes more flexible. This may force the model to not generalize new data not seen in the training set!

# Mathematics

The linear algebra is still the same as linear regression! The linear aspect refers to the coefficients we use to paint our picture.
The only difference between the models is that we transform the input data by adding polynomial terms (squaring/cubing/...) before applything the linear regression algorithm. This is what allows our model to fit a wider range of data patterns. ALl the underlying topics are the same.

# Polynomial Regression

### Importing Datset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv('../Datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values #Don't need position name since we have level
y = dataset.iloc[:, -1].values
```

### Training

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y) #didn't split into training and test set since we don't have enough data
poly = PolynomialFeatures(degree = 2) 
X_poly = poly.fit_transform(X) 
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

### Visualizing Linear and polynomial results

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.plot(X, poly_reg.predict(X_poly), color='green')
plt.plot(X)
plt.title('Linear Regression Model')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

This initial plot is showing us (in a very simple environment) how much better our polynomial fit is at predicting values. As discussed before, we can add more degrees to even fit the data closer!

### Visualizing Higher Degree polynomial

```python
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Polynomial Regression Model')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
